# Cubic equation $ax^3+bx^2+cx+d=0$

A cubic equation is an algebraic equation of the third degree, which means that the unknown variable (usually denoted as "$x$") is raised to the third power and may involve terms of the second degree, first degree, and constants.

To solve this problem, it is important to take into account that we are considering $a=1$ and $x_1,x_2$ and $x_3$ solutions, these three solutions are returned when we call the function ```solve_cubic(b,c,d)``` which has as parameters $b, c$ and $d$ being the coefficients of the equation.

# General cubic formula:

### Calculation of Coefficients $d_0$ and $d_1$

   - $d_0 = b^2 - 3c$: This coefficient is derived from the coefficients of the cubic equation and is crucial in determining the nature of the roots.
   - $d_1 = 2b^3 - 9bc + 27d$: $d_1$ is another coefficient that plays a significant role in the solution. It's important to note that $d_1$ is a symmetric polynomial in the roots.

### Calculation of $C_1$ and $C_2$:
   - $C_1 = \sqrt[3]{\frac{d_1 + \sqrt{d_1^2 - 4d_0^3}}{2}}$
   - $C_2 = \sqrt[3]{\frac{d_1 - \sqrt{d_1^2 - 4d_0^3}}{2}}$

   $C_1$ and $C_2$ are complex values used to calculate the roots of the cubic equation. The cubic root in these calculations can have multiple possible values due to the complex nature of the cubic equation.

### Checking Special Cases:
   - The method checks for special cases, such as when $d_1 = 2b^3$ or when $d_0$ and $d_1$ are both zero. These cases have particular solutions and do not follow the same calculation route as other cubic equations.

### Calculation of Roots $x_1$, $x_2$, and $x_3$:
   - If $d_1 = 2b^3$, a special formula is used: $x_1 = -\frac{b}{3} + \sqrt[3]{\frac{d}{2}} + \sqrt[3]{\frac{d}{2}}$. This case corresponds to a cubic equation that can be simplified into more straightforward terms.
   - In other cases, a choice is made between using $C_1$ or $C_2$ to calculate the roots $x_1$, $x_2$, and $x_3$ of the cubic equation. This is done considering the complex nature of the roots and selecting the appropriate formula.

### Approximation of Roots:
   - After calculating the roots, the code may use an approximation function to improve the accuracy of the solutions, especially when the roots are irrational or complex numbers.

### Return of Roots:
> Finally, the code returns the roots $x_1$, $x_2$, and $x_3$ as the result of solving the cubic equation.

This method is an advanced application of algebra and number theory that allows finding solutions to general cubic equations,

````python
import numpy as np
def cubic(b,c,d,x):
    """Calulate the value of the  qubic equation, v = x**3 + bx**2 + c*x +d

    Parameters
    ----------
    b, c, d  : float
       Coefficients

    Returns
    -------
    v : float """

    v = (x**3) + b*(x**2) + c*x + d

    if (str(type(v))=="" or str(type(v))==""):
      epsilon = 0 - v
      if (epsilon < 1e-8):
        v = 0.0
    else:
      real = v.real
      img = v.imag
      if (-real < 1e-8):
        real = 0.0
      if (-img < 1e-8):
        img = 0.0
      if (real == 0 and img == 0):
        v = 0.0
      else:
        v = complex(real, img)
    return v
````

````python
import numpy as np
import matplotlib.pyplot as plt

def cubic_eq(b, c, d, x):
    return x**3 + b*x**2 + c*x + d

def cubica(b, c, d, lx1=-10, lx2=10, num_points=400):
    x = np.linspace(lx1, lx2, num_points)
    y = cubic_eq(b, c, d, x)

    equation = f"y = x^3 + {b}x^2 + {c}x + {d}"

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=equation)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfica de una Ecuación Cúbica')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

cubica(4, -2, -3, -6, 2)
````

````python
import math
def solve_quad(b, c):
    """Solve a quadratic equation, x**2 + bx + c = 0.

    Parameters
    ----------
    b, c : float
       Coefficients

    Returns
    -------
    x1, x2 : float or complex
       Roots.  """

    # ... enter your code here ...
    discriminant = ((b**2)-(4*c))
    if discriminant == b**2:
      x1 = -b
      x2 = c/x1
    elif discriminant < 0:
      x1 = complex((-b/2), (math.sqrt(abs(discriminant))/2))
      x2 = complex((-b/2), (-math.sqrt(abs(discriminant))/2))
    else:
      x1 = (-b + (math.sqrt(discriminant)))/2
      x2 = (-b - (math.sqrt(discriminant)))/2

    return x1, x2
````

````python
import cmath
def aprox(a):
    real_part = a.real  # Obtiene la parte real del número complejo
    rounded_real_part = round(real_part)  # Redondea la parte real
    rounded_imag_part = round(a.imag, 10)

    # Crea un nuevo número complejo con la parte real redondeada y la parte imaginaria original
    rounded_complex = complex(rounded_real_part, rounded_imag_part)

    if ((abs(rounded_real_part-real_part)) < 1e-10) and abs(a.imag)>1e-10 :
      return rounded_complex
    elif abs(a.imag) <= 1e-10 and (abs(rounded_real_part-real_part)) < 1e-10:
      return rounded_real_part
    elif abs(a.imag) == 0 :

      return real_part
    else:
      return a

def solve_cubic(b, c,d):
    """Solve a cubic equation, x**3 + bx**2 + c*x +d = 0.

    Parameters
    ----------
    b, c,d : float
       Coefficients

    Returns
    -------
    x1, x2,x3 : float or complex
       Roots.  """
    d0 = b**2 -3*c
    d1 = 2*b**3-9*b*c+27*d
    C1=((d1+cmath.sqrt(d1**2-4*d0**3))/2)**(1/3)
    C2=((d1-cmath.sqrt(d1**2-4*d0**3))/2)**(1/3)
    ep=(-1+cmath.sqrt(-3))/2
    if c==d and c==0:
      print("c=0")
      x1 = 0
      x2 = 0
      x3 = -b
    elif d==0:
      print("d=0")
      x1 = 0
      x2= -b
      x3 = -c/b
    elif b>1e5 :
        print("Usa C1 grande")
        x1 = -(1/3) * (b + ep**0*C1 + d0/(ep**0*C1))
        x1 = aprox(x1)
        x2 = -(b+x1)
        x2 = aprox(x2)
        x3 = (x1**2+(b*x1))/d
        x3 = aprox(x3)
        #x1 es raiz real

    else:
      if d0==0 and d1==0:
        print("d0=0,d1=0")
        x1 = -b/3
        x1 = aprox(x1)
        x2 = -b/3
        x2 = aprox(x2)
        x3 = -b/3
        x3 = aprox(x3)
      elif C1 == 0:
        print("Usa C2")
        x1 = -(1/3) * (b + ep**0*C2 + d0/(ep**0*C2))
        x1 = aprox(x1)
        x2 = -(1/3) * (b + ep**1*C2 + d0/(ep**1*C2))
        x2 = aprox(x2)
        x3 = -(1/3) * (b + ep**2*C2 + d0/(ep**2*C2))
        x3 = aprox(x3)
      else:
        print("Usa C1","  d1= ",d1, " == ", 2*b**3)
        x1 = -(1/3) * (b + ep**0*C1 + d0/(ep**0*C1))
        x1 = aprox(x1)
        x2 = -(1/3) * (b + ep**1*C1 + d0/(ep**1*C1))
        x2 = aprox(x2)
        x3 = -(1/3) * (b + ep**2*C1 + d0/(ep**2*C1))
        x3 = aprox(x3)
    return x1,x2,x3

x1, x2, x3 = solve_cubic(-1e6,1.0,0.0)

print(x1,x2,x3)
````
